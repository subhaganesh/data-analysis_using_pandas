# apply method with row values

```python
bond=pd.read_csv('jamesbond.csv',index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def good_movie(row):
    actor= row[1]
    budget=row[4]
    
    if actor == 'Roger Moore':
        return 'best'
    elif actor == 'Daniel Craig' and budget >40:
        return 'ok'
    else:
        return 'no clue'
    
bond.apply(good_movie,axis=1)
```
