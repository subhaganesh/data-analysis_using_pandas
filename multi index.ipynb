# the pivot method

```python
sales=pd.read_csv('salesmen.csv',parse_dates=['Date'])
sales['Salesman']=sales['Salesman'].astype('category')
sales.head(3)
```

```python
len(sales)
```

```python
sales['Salesman'].value_counts()
```

```python
sales.pivot(index='Date',columns='Salesman',values='Revenue')
```

# pivot_table method

```python
food=pd.read_csv('foods.csv')
food.head(3)
```

```python
food.pivot_table(values='Spend',index=['Gender','Item'],columns='City',aggfunc='sum')
```

```python
pd.pivot_table(data=food,values='Spend',index=['Gender','Item'],columns='City',aggfunc='sum')
```
